```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn import linear_model
%matplotlib inline
```

```python
# Import listings.csv and load it as pandas dataframe
df_listings = pd.read_csv("listings.csv")
```

```python
# Print out the head of the listings data
df_listings.head()
```

```python
# List out all the columns in the listings
list(df_listings.columns)
```

```python
# Check correlation matrix
corr = df_listings.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
f, ax = plt.subplots(figsize=(20, 9))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
# Draw a heatmap 
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .70})
```

```python
# Remove unwanted characters and convet into float
df_listings.price = df_listings.price.replace('[\$,]', '', regex=True).astype(float)
```

```python
# Convert columns into datetime
df_listings['host_since'] = pd.to_datetime(df_listings['host_since'])
```

```python
# Group the data by the column 'host_since'
result = df_listings.groupby(df_listings['host_since'].map(lambda x: x.year)).id.count()
year = pd.DataFrame(result)
```

```python
# Rename the 'id' column to listings
year.rename(columns={'id':'listings'}, inplace=True)
year = pd.DataFrame(year).reset_index()
```

```python
# Create a column with cumsum of the listings by year
year['total listings']= year['listings'].cumsum(axis=0)
```

```python
# Convert'host_since' to int
year['host_since'] = year['host_since'].astype(int)
# Remnaming the column 'Host_Since' to 'year'
year.rename(columns={'host_since':'year'}, inplace=True)
```

```python
#Cambridge,MA Airbnb listings over year
sns.set(style="darkgrid")
fig, ax = plt.subplots()
ax = sns.barplot(x="year", y="total listings", data=year, color='Coral')
ax.set_title('Airbnb Listings in Cambridge')
```

```python
#New Airbnb Listings in Cambridge, MA
sns.set(style="darkgrid")
fig, ax = plt.subplots()
ax = sns.barplot(x='year', y="listings", data=year, color="coral")
ax.set_title('New Airbnb Listings in Cambridge')
```

```python
#Room Type in Cambridge, MA
sns.set(style="darkgrid")
fig, ax = plt.subplots()
# Title
ax.axes.set_title("Room Type in Cambridge, MA")
# Countplot for 'room_type' and their count
ax = sns.countplot(y='room_type',data=df_listings,order=df_listings['room_type'].value_counts().index, palette="Set3")
# X and Y Lables
ax.set_xlabel('Number of Listings')
ax.set_ylabel('Room Type')
```

```python
#Top neighbourhoods in Cambridge, MA
sns.set(style="darkgrid")
fig, ax = plt.subplots()

ax.axes.set_title("Top neighbourhoods in Cambridge, MA")
# Countplot for 'neighbourhood' and their count
ax = sns.countplot(y='neighbourhood',data=df_listings, order = df_listings['neighbourhood'].value_counts().iloc[:20].index, palette="Set3")
# X and Y labels
ax.set_xlabel('Number of Listings')
ax.set_ylabel('Neighbourhood')
```

```python
#load calendar data
calendar = pd.read_csv("calendar.csv")
calendar.head(10)
```

```python
#Print columns information 
calendar.info()
```

```python
#Remove $ sign on the price column and convert to numeric value
calendar['price'] =calendar['price'].apply(lambda x: str(x).replace('$',''))
calendar['price'] = pd.to_numeric(calendar['price'], errors = 'coerce')
df_calendar = calendar.groupby('date')[["price"]].sum()
df_calendar['mean'] = calendar.groupby('date')[["price"]].mean()
df_calendar.columns = ['Total Price', 'Avg Price']
df_calendar.head(10)
```

```python
#Set date as index
df_calendar2 = calendar.set_index("date")
df_calendar2.index = pd.to_datetime(df_calendar2.index)
df_calendar2 = df_calendar2[['price']].resample('M').mean()
df_calendar2.head(10)
```

```python
#Summary statistics
print(df_calendar2.describe())
```

```python
# Price by Month Line Chart
ax=df_calendar2.plot(figsize=(12,5), color='blue')
ax.set_title('Price by Month')
ax.set_xlabel('Date')
ax.set_ylabel('Price')
plt.show()
```
